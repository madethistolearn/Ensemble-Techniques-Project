```python
import matplotlib.pyplot as plt

def create_pie_charts(dataframe):  
    # Getting the list of categorical columns
    categorical_columns = dataframe.select_dtypes(include=['object']).columns

    for column_name in categorical_columns:
        # Calculating value counts and percentages for each category
        value_counts = dataframe[column_name].value_counts()
        labels = value_counts.index
        sizes = value_counts.values
        percentages = (sizes / sizes.sum()) * 100

        # Creating a pie chart
        plt.figure(figsize=(6, 6))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title(f'Percentage Distribution of {column_name}')
        plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is drawn as a circle.

        # Displaying the pie chart
        plt.show()

create_pie_charts(df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Specifying the columns for binary encoding
binary_encode_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']

# Creating a LabelEncoder instance
label_encoder = LabelEncoder()

# Binary encoding for binary categorical features
for column in binary_encode_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

### Using One-Hot Encoding on the following features as they contain multiple categories:

##### MultipleLines (Three categories: No phone service, Yes, No)
##### InternetService (Three categories: No, DSL, Fiber Optic)
##### OnlineSecurity (Three categories: No internet service, Yes, No)
##### OnlineBackup (Three categories: No internet service, Yes, No)
##### DeviceProtection (Three categories: No internet service, Yes, No)
##### TechSupport (Three categories: No internet service, Yes, No)
##### StreamingTV (Three categories: No internet service, Yes, No)
##### StreamingMovies (Three categories: No internet service, Yes, No)
##### Contract (Three categories: One year, Two year, Month-to-month)
##### PaymentMethod (Four categories: Credit card (automatic), Bank transfer (automatic), Mailed check, Electronic check)

```python
from sklearn.preprocessing import OneHotEncoder

# Specifying the columns for one-hot encoding
one_hot_encode_columns = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
                          'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
                          'Contract', 'PaymentMethod']

# Creating a OneHotEncoder instance
one_hot_encoder = OneHotEncoder(sparse_output=False, drop='first', dtype=int)

# Fitting and transforming the specified columns
df_encoded = one_hot_encoder.fit_transform(df[one_hot_encode_columns])

# Getting the feature names for the one-hot encoded columns
feature_names = one_hot_encoder.get_feature_names_out(input_features=one_hot_encode_columns)

# Creating a DataFrame with the one-hot encoded columns
df_encoded_df = pd.DataFrame(df_encoded, columns=feature_names)

# Concatenate the one-hot encoded DataFrame with the original DataFrame
df = pd.concat([df, df_encoded_df], axis=1)

# Drop the original columns that were one-hot encoded
df.drop(one_hot_encode_columns, axis=1, inplace=True)
```

```python
# Using X to represent the features or independent variables in the dataset
X = df.drop('Churn', axis=1) #creating the X variable by using the drop method on the DataFrame df
##The X variable will contain all the feature columns except the target column
#We use Y to represent the target variable or the dependent variable that the model is trying to predict or classify
Y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split #importing train_test_split from sklearn 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
#The above line of code takes the feature data X and target data Y, and randomly splits them into training and testing sets
#The training sets (X_train and Y_train) will be used to train the machine learning model
#The testing sets (X_test and Y_test) will be used to evaluate the model's performance.
```

```python
# Importing ‘StandardScaler’ from ‘Sklearn’ for Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### A. Train a model using Decision tree and check the performance of the model on train and test data

```python
from sklearn.tree import DecisionTreeClassifier

dTree = DecisionTreeClassifier(criterion = 'gini', random_state=1)
dTree.fit(X_train, Y_train)
```

```python
#checking performance
print(dTree.score(X_train, Y_train))
print(dTree.score(X_test, Y_test))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Getting predictions from the model
Y_predict = dTree.predict(X_test)

# Converting string labels to integer labels for the confusion matrix
Y_test_int = [1 if label == "Yes" else 0 for label in Y_test]
Y_predict_int = [1 if label == "Yes" else 0 for label in Y_predict]

# Creating the confusion matrix using integer labels
cm = confusion_matrix(Y_test_int, Y_predict_int)

# Create a DataFrame for the confusion matrix
df_cm = pd.DataFrame(cm, index=["No", "Yes"], columns=["No", "Yes"])

# Plot the confusion matrix
plt.figure(figsize=(7, 5))
sns.heatmap(df_cm, annot=True, fmt='g')
```

### B. Use grid search and improve the performance of the Decision tree model , check the performance of the model on train and test data , provide the differences observed in performance in Q3.a and Q3.b

```python
from sklearn.model_selection import GridSearchCV

# Creating a Decision Tree classifier
dt_classifier = DecisionTreeClassifier()

# Defining the hyperparameter grid to search through
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Creating a GridSearchCV instance
grid_search = GridSearchCV(dt_classifier, param_grid, cv=5, scoring='accuracy')

# Fitting the grid search to your training data
grid_search.fit(X_train, Y_train)

# Getting the best parameters and the best estimator
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

# Evaluating the best estimator on your test data
test_score = best_estimator.score(X_test, Y_test)

# Printing the best parameters and the test score
print("Best Parameters:", best_params)
print("Test Accuracy:", test_score)
```
